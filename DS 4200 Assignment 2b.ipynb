```python
import matplotlib.pyplot as plt
import matplotlib.cbook as cbook

import numpy as np
import pandas as pd

dataframe1 = pd.read_excel('/Users/ashwinrupakula/Downloads/flowers.xlsx')
dataframe1
```

Question 3 <br />

The data describes different categories of species of plants and when they were discovered or found <br />
Categorical :- Kingdom, Geneus, Species, Habitat. 
Ordinal :- Hosmer, Eaton, Primack.
Quantitative :- year, First Flower Date.


```python
dataframe1['number_of_genus'] = None
for name, group in dataframe1.groupby(by=['Habitat']):
    dataframe1.loc[dataframe1.Habitat == name, ['number_of_genus']] = len(list(group.Genus.unique()))

dataframe1.head(5)
```

```python
#Question 4
habitat = dataframe1['Habitat']
no_of_genus = dataframe1['number_of_genus']
plt.bar(habitat, no_of_genus)
```

```python
plt.plot(habitat, no_of_genus)
```

#Question 5 <br />
The more appropriate way to represent the data above is by using a bar graph. This is because the graph provides a more accurate representation of a total, rather than showing some form of progression

```python
#Question 6
plt.bar(habitat, no_of_genus)
plt.title("How many Genuses does each Habitat have")
plt.xlabel("Habitat")
plt.ylabel("Number of Genus per Habitat")
```

```python
#Question 7
plt.figure()
for name, group in dataframe1.groupby(by=['Genus']):
    plt.plot(group['year'], group['firstFlowerDate'], label=name)

plt.title("First flower date for each genus over time")
plt.xlabel("Year")
plt.ylabel("First Flower Date")
plt.legend(bbox_to_anchor=(1.05,0.6))
plt.show()
```
