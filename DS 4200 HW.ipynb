```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
 
# --- FORMAT 1
 
# Your x and y axis
x=range(1,6)
y=[ [1,3,5,7,20], [2,6,7,14,17], [2,6,5,4,6] ]
 
# Basic stacked area chart.
plt.stackplot(x,y, labels=['x','y','z'])
plt.legend(loc='upper left')
plt.show()
 
```

I choose to implement this graph as it is very important to show the evolution of the value of several groups on the same graphic. This leads to more observations being viewed and allows for a comparison between multiple variables
